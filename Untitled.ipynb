```python
final['Locality_1'] = ''
final['Locality_2'] = ''
final['Locality_3'] = ''

locality_split= final['Locality'].str.split(",", n = -1, expand = True) 


final['Locality_1'] = locality_split[0]
final['Locality_2'] = locality_split[1]
final['Locality_3'] = locality_split[2]

final
```

```python
final.isnull().sum()
```

```python
final.isnull().sum(axis=1)
```

```python
final_india = final[final['Country'] == 'India']
```

```python
top = final_india.groupby(['Cuisines','City'])['Aggregate rating'].sum().reset_index()
```

```python
top_df = pd.DataFrame(top)
top_df = top_df[top_df['City'] == 'New Delhi']
```

```python
plt.figure(figsize=(30,15))
sns.barplot(y='Aggregate rating', x="Cuisines", data=top_df.nlargest(30,'Aggregate rating'))
plt.title('New Delhi Cuisines and Rating - Top 30')
plt.xticks(rotation=90)
plt.legend(top_df.nlargest(30, 'Aggregate rating')['Cuisines'])
plt.show()
```

```python
all_other = final[final['Country'] != 'India'].groupby(['Cuisines','Country'])['Aggregate rating'].count().reset_index().sort_values(by='Aggregate rating', ascending = False)
```

```python
US = all_other[all_other['Country'] == 'United States'].sort_values(by='Aggregate rating', ascending = False)
US
```

```python
plt.figure(figsize=(27,18))
sns.barplot(y='Aggregate rating', x="Cuisines", data=US.nlargest(30,'Aggregate rating'))
plt.title('New Delhi Cuisines and Rating - Top 30')
plt.xticks(rotation=90)
plt.legend(US.nlargest(30, 'Aggregate rating')['Cuisines'])
plt.show()
```
